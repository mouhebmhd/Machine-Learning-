<font size='3'>Maintenant, examinons les frais par r√©gion.</font>


```python
# Importation des biblioth√®ques n√©cessaires pour la visualisation
import seaborn as sns
import matplotlib.pyplot as plt

# Supposons que 'df' soit votre DataFrame
# Groupement des charges par r√©gion, puis calcul de la somme des charges pour chaque r√©gion
# Les valeurs sont ensuite tri√©es par ordre croissant
charges = df['charges'].groupby(df.region).sum().sort_values(ascending=True)

# Cr√©ation d'une figure avec un seul sous-graphe de taille 8x6 pouces
f, ax = plt.subplots(1, 1, figsize=(8, 6))

# Tracer un diagramme √† barres avec les r√©gions sur l'axe x et les charges correspondantes sur l'axe y
# La palette de couleurs utilis√©e est 'Blues'
sns.barplot(x=charges.index, y=charges.values, palette='Blues', ax=ax)

# D√©finir le titre du graphique
ax.set_title('Distribution of Charges')

# Afficher le graphique
plt.xticks(rotation=45)  # Rotation des √©tiquettes de l'axe x si n√©cessaire pour une meilleure lisibilit√©
plt.show()
```

<font size='3'>Ainsi, en g√©n√©ral, les frais m√©dicaux les plus √©lev√©s se trouvent dans le Sud-Est et les plus bas dans le Sud-Ouest. En tenant compte de certains facteurs (sexe, tabagisme, avoir des enfants), voyons comment cela varie selon les r√©gions.</font>


```python
# Cr√©ation d'une figure avec un seul sous-graphe de taille 12x8 pouces
f, ax = plt.subplots(1, 1, figsize=(12, 8))

# Tracer un diagramme √† barres avec les r√©gions sur l'axe x et les charges sur l'axe y
# L'argument 'hue' permet de distinguer les charges selon le sexe ('sex') des individus
# La palette de couleurs utilis√©e est 'cool'
ax = sns.barplot(x='region', y='charges', hue='sex', data=df, palette='cool')
```

```python
# Cr√©ation d'une figure avec un seul sous-graphe de taille 12x8 pouces
f, ax = plt.subplots(1, 1, figsize=(12, 8))

# Tracer un diagramme √† barres avec les r√©gions sur l'axe x et les charges sur l'axe y
# L'argument 'hue' permet de distinguer les charges selon le nombre d'enfants ('children')
# La palette de couleurs utilis√©e est 'Set1', qui offre une gamme de couleurs vives
ax = sns.barplot(x='region', y='charges', hue='children', data=df, palette='Set1')
```
